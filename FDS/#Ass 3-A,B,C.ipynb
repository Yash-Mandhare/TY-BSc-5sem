```python

import pandas as pd
df=pd.read_csv(r'C:\Users\practical\Downloads\Data.csv')
print('DataFrame\n------------------------------------------\n')
print(df)
print('\n\nStatistical description\n---------------------------------------\n')
print(df.describe())
print('\nNumber of rows8columns:',df.shape)
print(df.head(3))
```

```python
# Q2.
agemean=df['Age'].mean()
salarymean=df['Salary'].mean()
df['Age'].fillna(agemean,inplace=True)
df['Salary'].fillna(salarymean,inplace=True)
print("\n\nMissing values of age and salary replaced with mean value:")
df
```

```python
# Q3.
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
he=OneHotEncoder()
enc_df=pd.DataFrame(he.fit_transform(df[['Country']]).toarray())
df=df.join(enc_df)
print('\n\nOne Hot encoding on Country column:\n------------------------------------\n')
print(df)
labelencoder=LabelEncoder()
df['Purchased']=labelencoder.fit_transform(df['Purchased'])
print('\nlabel encoding on purchased column')
print(df)
```

```python
# Set B
# Q1.
import pandas as pd
import scipy.stats as s
from sklearn import preprocessing
df=pd.read_csv(r'C:\Users\practical\Downloads\winequality-red (1).csv')
df
```

```python
# Q2.
print("\n\nData scaled between 0 to 1")
data_scaler=preprocessing.MinMaxScaler(feature_range=(0,1))
data_scaled=data_scaler.fit_transform(df)
print("\nMin Max scaled data\n---------------------------------------------")
print(data_scaled.round(2))
```

```python
# Q3.
print("Original data\n----------------------------------------------")
print(df)
print("\nInitial mean:",s.tmean(df).round(2))
print("\nInitial standard deviation\n----------------------------------------------")
print(round(df.std(),2))
df_scaled=preprocessing.scale(df)
df_scaled.mean(axis=0)
df_scaled.std(axis=0)
print("\nStandardize data\n",df_scaled.round(2))
print("\nScaled mean:",s.tmean(df_scaled).round(2))
print("\nscaled standard deviation:",round(df_scaled.std(),2))
```

```python
# Q4
dn=preprocessing.normalize(df,norm='l1')
print("\n L1 normalized data\n--------------------------------------------")
print(dn.round(2))
```

```python
# Q5.
data_binarized=preprocessing.Binarizer(threshold=5).transform(df)
print("\nBinarized data\n----------------------------------------")
print(data_binarized)
```

```python
# Set C
# Q1.
import pandas as pd
import numpy as np
df=pd.read_csv(r'Student_bucketing.csv')
print("\nOriginal dataset\n-------------------------------------")
print(df)

m1=min(df["marks"])
m2=max(df["marks"])

bins=np.linspace(m1,m2,6)
names=["Poor","Below_average","Average","Above_average","Excellent"]
df["marks_bin"]=pd.cut(df["marks"],bins,labels=names,include_lowest=True)
print("\nBinned dataset\n---------------------------------------------")
df.sample(15)
```
