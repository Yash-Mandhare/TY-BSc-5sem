```python
# A Write a Python program to create a Pie plot to get the frequency of the three species of
# the Iris data (Use iris.csv)

import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the Iris data from the CSV file
data = pd.read_csv('iris.csv')

# Step 2: Group the data by the 'species' column and count the frequency of each species
species_counts = data['species'].value_counts()

# Step 3: Create a pie plot to visualize the frequency of species
plt.figure(figsize=(6, 6))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Frequency of Iris Species')
plt.axis('equal')  # Equal aspect ratio ensures that the pie chart is circular.

# Step 4: Display the pie plot
plt.show()
```

```python
# Write a Python program for Handling Missing Value. Replace missing value of salary,
# age column with mean of that column.(Use Data.csv file).

import pandas as pd

# Step 1: Read the data from the CSV file
data = pd.read_csv('Data.csv')

# Step 2: Calculate the mean of the 'salary' and 'age' columns
mean_salary = data['Salary'].mean()
mean_age = data['Age'].mean()

# Step 3: Replace missing values with the respective means
data['Salary'].fillna(mean_salary, inplace=True)
data['Age'].fillna(mean_age, inplace=True)

# Step 4: Save the updated data to a new CSV file (if needed)
data.to_csv('Data_with_missing_values_handled.csv', index=False)

# Step 5: Display the updated data
print("Data with missing values replaced by means:")
print(data)

# If you want to save the updated data to a new CSV file, you can uncomment the Step 4 line.

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Read the Iris data from the CSV file
data = pd.read_csv('iris.csv')

# Step 2: Create box plots to visualize the distribution of features across species
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed

# Sepal Length
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Sepal Length')

# Sepal Width
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=data)
plt.title('Sepal Width')

# Petal Length
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=data)
plt.title('Petal Length')

# Petal Width
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=data)
plt.title('Petal Width')

plt.tight_layout()

# Step 3: Display the box plots
plt.show()
```

```python
# Generate a random array of 50 integers and display them using a line chart, scatter
# plot, histogram and box plot. Apply appropriate color, labels and styling options.

import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate a random array of 50 integers
np.random.seed(42)  # Setting a random seed for reproducibility
random_array = np.random.randint(1, 101, size=50)

# Step 2: Create a figure with multiple subplots
plt.figure(figsize=(12, 6))

# Line Chart
plt.subplot(2, 2, 1)
plt.plot(random_array, color='blue', marker='o', linestyle='-', label='Random Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Line Chart')
plt.legend()

# Scatter Plot
plt.subplot(2, 2, 2)
plt.scatter(range(len(random_array)), random_array, color='green', marker='o', label='Random Data')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Scatter Plot')
plt.legend()

# Histogram
plt.subplot(2, 2, 3)
plt.hist(random_array, bins=10, color='orange', edgecolor='black', label='Random Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.legend()

# Box Plot
plt.subplot(2, 2, 4)
plt.boxplot(random_array, vert=False, labels=['Random Data'], patch_artist=True)
plt.xlabel('Value')
plt.title('Box Plot')

plt.tight_layout()

# Step 3: Display the plots
plt.show()
```
