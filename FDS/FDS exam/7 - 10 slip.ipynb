```python
# Write a Python program to perform the following tasks :
# a. Apply OneHot coding on Country column.


import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Load the CSV data
data = pd.read_csv('Data-Country.csv')

# Apply One-Hot Encoding to the "Country" column
encoder = OneHotEncoder(sparse=False)
encoded_country = encoder.fit_transform(data[['Country']])

# Create a DataFrame for the one-hot encoded "Country" column
encoded_country_df = pd.DataFrame(encoded_country, columns=encoder.get_feature_names_out(['Country']))

# Concatenate the one-hot encoded data with the original DataFrame
data_encoded = pd.concat([data, encoded_country_df], axis=1)

# Print the modified DataFrame
print(data_encoded)
```

```python
# b. Apply Label encoding on purchased column

import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the CSV data
data = pd.read_csv('Data-Country.csv')

# Apply Label Encoding to the "Purchased" column
label_encoder = LabelEncoder()
data['Purchased'] = label_encoder.fit_transform(data['Purchased'])

# Print the modified DataFrame
print(data)
```

```python
# Write a program in python to perform following task : 
# Standardizing Data (transform them into a standard Gaussian distribution with a mean
# of 0 and a standard deviation of 1) (Use winequality-red.csv)

import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the CSV data
data = pd.read_csv('winequality-red.csv')

# Extract the features (independent variables) you want to standardize
features = data.drop('quality', axis=1)

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to your data and transform it
standardized_features = scaler.fit_transform(features)

# Create a DataFrame with the standardized features
standardized_data = pd.DataFrame(standardized_features, columns=features.columns)

# Print the standardized data
print(standardized_data)

```

```python
# Generate a random array of 50 integers and display them using a line chart, scatter plot.
# Apply appropriate color, labels and styling options.

import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of 50 integers
np.random.seed(0)
random_data = np.random.randint(0, 100, 50)

# Create an array of x values (indices) for the data points
x_values = np.arange(1, 51)

# Create a figure and two subplots for the line chart and scatter plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Line Chart
ax1.plot(x_values, random_data, marker='o', linestyle='-', color='blue', label='Line Chart')
ax1.set_title('Line Chart')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')
ax1.legend()

# Scatter Plot
ax2.scatter(x_values, random_data, c='red', marker='x', label='Scatter Plot')
ax2.set_title('Scatter Plot')
ax2.set_xlabel('Index')
ax2.set_ylabel('Value')
ax2.legend()

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Create two lists, one representing subject names and the other representing marks
# obtained in those subjects. Display the data in a pie chart.

import matplotlib.pyplot as plt

# Create two lists representing subject names and marks
subject_names = ["Math", "Science", "History", "English", "Art"]
marks_obtained = [90, 85, 78, 92, 70]

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(marks_obtained, labels=subject_names, autopct='%1.1f%%', startangle=140, shadow=True)

# Add a title
plt.title("Marks Obtained in Subjects")

# Display the pie chart
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```

```python
# Write a python program to compute sum of Manhattan distance between all pairs of
# points.

from itertools import combinations

# Function to calculate Manhattan distance between two points
def manhattan_distance(point1, point2):
    return abs(point1[0] - point2[0]) + abs(point1[1] - point2[1])

# List of points (coordinates)
points = [(1, 2), (3, 4), (5, 6), (7, 8)]

# Initialize the total Manhattan distance
total_distance = 0

# Generate all pairs of points
point_pairs = combinations(points, 2)

# Calculate the sum of Manhattan distances between all pairs
for pair in point_pairs:
    total_distance += manhattan_distance(pair[0], pair[1])

# Print the total Manhattan distance
print("Sum of Manhattan distances between all pairs of points:", total_distance)
```
