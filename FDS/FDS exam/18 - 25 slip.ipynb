```python
# Write a Python program to create box plots to see how each feature i.e. Sepal Length,
# Sepal Width, Petal Length, Petal Width are distributed across the three species. (Use iris.csv
# dataset)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = pd.read_csv('iris.csv')  # Make sure the path is correct

# Create box plots using Seaborn
plt.figure(figsize=(10, 6))

# Box plot for Sepal Length across species
plt.subplot(2, 2, 1)
sns.boxplot(x='species', y='sepal_length', data=iris)

# Box plot for Sepal Width across species
plt.subplot(2, 2, 2)
sns.boxplot(x='species', y='sepal_width', data=iris)

# Box plot for Petal Length across species
plt.subplot(2, 2, 3)
sns.boxplot(x='species', y='petal_length', data=iris)

# Box plot for Petal Width across species
plt.subplot(2, 2, 4)
sns.boxplot(x='species', y='petal_width', data=iris)

plt.tight_layout()
plt.show()
```

```python
# Generate a random array of 50 integers and display them using a line chart, scatter plot,
# histogram and box plot. Apply appropriate color, labels and styling options.


import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of 50 integers
np.random.seed(42)  # Set a random seed for reproducibility
random_data = np.random.randint(1, 100, 50)  # Generating 50 random integers between 1 and 100

# Line Chart
plt.figure(figsize=(12, 3))
plt.subplot(141)
plt.plot(random_data, color='b', marker='o')
plt.title('Line Chart')
plt.xlabel('Data Index')
plt.ylabel('Value')

# Scatter Plot
plt.subplot(142)
plt.scatter(range(len(random_data)), random_data, color='g', marker='x')
plt.title('Scatter Plot')
plt.xlabel('Data Index')
plt.ylabel('Value')

# Histogram
plt.subplot(143)
plt.hist(random_data, bins=10, color='r', edgecolor='k')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Box Plot
plt.subplot(144)
plt.boxplot(random_data, vert=False, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Value')

# Adding style
plt.tight_layout()
plt.show()
```

```python
# Add two outliers to the above data and display the box plot.

import numpy as np
import matplotlib.pyplot as plt

# Generate a random array of 50 integers
np.random.seed(42)  # Set a random seed for reproducibility
random_data = np.random.randint(1, 100, 50)  # Generating 50 random integers between 1 and 100

# Add two outliers
outliers = [150, 160]
random_data = np.append(random_data, outliers)

# Box Plot with Outliers
plt.figure(figsize=(6, 4))
plt.boxplot(random_data, vert=False, patch_artist=True, flierprops={'marker': 'o', 'markerfacecolor': 'r', 'markersize': 8})
plt.title('Box Plot with Outliers')
plt.xlabel('Value')

# Adding style
plt.show()
```

```python
# Import dataset iris.csv. Write a Python program to create a Bar plot to get the
# frequency of the three species of the Iris data.


import pandas as pd
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = pd.read_csv('iris.csv')  # Make sure the path is correct

# Count the frequency of each species
species_counts = iris['species'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
species_counts.plot(kind='bar', color='skyblue')
plt.title('Frequency of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')

# Display the plot
plt.show()
```

```python
# Write a Python program to create a histogram of the three species of the Iris data.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = pd.read_csv('iris.csv')  # Make sure the path is correct

# Create histograms for each species
plt.figure(figsize=(10, 6))

# Histogram for Sepal Length
plt.subplot(2, 2, 1)
sns.histplot(iris, x='sepal_length', hue='species', kde=True, common_norm=False)
plt.title('Sepal Length')

# Histogram for Sepal Width
plt.subplot(2, 2, 2)
sns.histplot(iris, x='sepal_width', hue='species', kde=True, common_norm=False)
plt.title('Sepal Width')

# Histogram for Petal Length
plt.subplot(2, 2, 3)
sns.histplot(iris, x='petal_length', hue='species', kde=True, common_norm=False)
plt.title('Petal Length')

# Histogram for Petal Width
plt.subplot(2, 2, 4)
sns.histplot(iris, x='petal_width', hue='species', kde=True, common_norm=False)
plt.title('Petal Width')

plt.tight_layout()
plt.show()
```

```python
# Rescaling: Normalised the dataset using MinMaxScaler class

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the Iris dataset
iris = pd.read_csv('iris.csv')  # Make sure the path is correct

# Select the columns you want to normalize (in this case, the numerical features)
numerical_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create a MinMaxScaler instance
scaler = MinMaxScaler()

# Fit the scaler to the data and transform the data
iris[numerical_features] = scaler.fit_transform(iris[numerical_features])

# Display the normalized dataset
print(iris)
```

```python
# Standardizing Data (transform them into a standard Gaussian distribution with a mean of 0 and
# a standard deviation of 1)


import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the Iris dataset
iris = pd.read_csv('iris.csv')  # Make sure the path is correct

# Select the columns you want to standardize (in this case, the numerical features)
numerical_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create a StandardScaler instance
scaler = StandardScaler()

# Fit the scaler to the data and transform the data
iris[numerical_features] = scaler.fit_transform(iris[numerical_features])

# Display the standardized dataset
print(iris)
```

```python
# Normalizing Data ( rescale each observation to a length of 1 (a unit norm). For this, use the
# Normalizer class.)

import pandas as pd
from sklearn.preprocessing import Normalizer

# Load the Iris dataset
iris = pd.read_csv('iris.csv')  # Make sure the path is correct

# Select the columns you want to normalize (in this case, the numerical features)
numerical_features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Create a Normalizer instance
normalizer = Normalizer()

# Transform the data to normalize each observation to a length of 1
iris[numerical_features] = normalizer.transform(iris[numerical_features])

# Display the normalized dataset
print(iris)
```

```python
# Binarizing Data using we use the Binarizer class (Using a binary threshold, it is possible
# to transform our data by marking the values above it 1 and those equal to or below it, 0)

import pandas as pd
from sklearn.preprocessing import Binarizer

# Load the Iris dataset
iris = pd.read_csv('iris.csv')  # Make sure the path is correct

# Select the column you want to binarize (e.g., 'sepal_length' in this example)
feature_to_binarize = 'sepal_length'

# Define a binary threshold (values above this threshold will be set to 1, others to 0)
threshold = 5.5

# Create a Binarizer instance with the specified threshold
binarizer = Binarizer(threshold=threshold)

# Transform the data to binarize the selected feature
iris[feature_to_binarize] = binarizer.transform(iris[[feature_to_binarize]])

# Display the binarized dataset
print(iris)
```
